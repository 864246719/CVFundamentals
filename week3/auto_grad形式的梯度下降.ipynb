# 当x是一个标量时

```python
print("%s%s"%("-"*20,"当x是1x1的时候"))
x = torch.ones(1,1,requires_grad = True)
y = (x+2)*(x+2)*3
```

```python
y.backward()
```

#用autograd来求解我们的模型 

```python
x = torch.randn(1,10)
```

```python
w = torch.ones(10,1,requires_grad=True)
```

```python
y = x.mm(w)
```

```python
print(y)
```

```python
loss = (y-8)*(y-8)
```

```python
print(loss)
```

```python
loss.backward()
```

```python
print(w.grad)
```

```python
w.data = w.data-w.grad*0.1
```

```python
print(x.mm(w))
```

```python
y = x.mm(w)
```

```python
loss=(y-8)*(y-8)
```

```python
loss.backward()
```

```python
w.data = w.data-w.grad*0.1
```

```python
print(x.mm(w))
```

```python
y = x.mm(w)
loss = (y-8)*(y-8)
print("y=%s,loss=%s"%(y.data,loss.data))
loss.backward()
w.data = w.data-w.grad*0.01
w.grad.data = w.grad.data-w.grad.data
print(w.grad)
```
